```python
import re
def remove_tags(text):
    text = re.sub('</?[^>]+>', '', text)
    # Вставка недостающих пробелов
    for item in (re.findall('[^т]([\.|\!|\,|\?]\w)', text)):
        text = re.sub(f'\{item[0]}{item[1]}', f'{item[0]} {item[1]}', text)
    return text
```

```python
# ваш код
# mystem
from pymystem3 import Mystem
mystem = Mystem()
words_analized = mystem.analyze(text)
text_lemmatized_mystem = [parse['analysis'][0]['lex'] for parse in words_analized if parse.get('analysis')]
```

```python
#pymorphy
from pymorphy2 import MorphAnalyzer
morph = MorphAnalyzer()
tokenized_text = [words_analized[i]['text'] for i in range(len(words_analized)) if words_analized[i].get('analysis')] 
next_words_analized = [morph.parse(token) for token in tokenized_text]
text_lemmatized_pymorphy = [next_words_analized[i][0].normal_form for i in range(len(next_words_analized))]
```

```python
text_lemmatized_pymorphy
```

```python
# Посмотрим на разницу
count = 0
for i in range(len(text_lemmatized_mystem)):
    if text_lemmatized_mystem[i] != text_lemmatized_pymorphy[i]:
        print(f'Токен: {tokenized_text[i]}, Mystem: {text_lemmatized_mystem[i]}, Pymorphy: {text_lemmatized_pymorphy[i]}')
    else:
        count += 1
print(f'Одинаковый разбор слова: {(count/len(text_lemmatized_mystem))*100} %')
```

## 6*. Дизамбигуация в Mystem (2 балла)

Одна из фичей майстема - контекстная дизамбигуация (т.е. определение правильного разбора в зависимости от контекста).

Запустите mystem на очищенном тексте с дизамбигуацией и без и проверьте есть ли какие-то различия.

```python
# ваш код
# Текст со снятой омонимией
# В консоли .\mystem.exe hw_text.txt hw_text_parsed_disamb.txt -c -s -d --format json
disamb_mystem = [json.loads(line) for line in open("C:\my_stem\hw_text_parsed_disamb.txt", encoding = 'utf-8')]
```

```python
# Из прошлого номера - текст без снятой омонимии, в консоли .\mystem.exe hw_text.txt hw_text_parsed.txt -c -s --format json
text_mystem
```

```python
lemmas_text_mystem = []
for i in text_mystem:
    lemmas_in_paragraph = [parse['analysis'][0]['lex'] for parse in i if parse.get('analysis')]
    lemmas_text_mystem.append(lemmas_in_paragraph)
lemmas_disamb_mystem = []
for j in disamb_mystem:
    lemmas_in_paragraph = [parse['analysis'][0]['lex'] for parse in j if parse.get('analysis')]
    lemmas_disamb_mystem.append(lemmas_in_paragraph)
```

```python
lemmas_text_mystem
```

```python
lemmas_disamb_mystem
```

```python
# Различия
for i in range(len(lemmas_text_mystem)):
    for j in range(len(lemmas_text_mystem[i])):
        if lemmas_text_mystem[i][j] != lemmas_disamb_mystem[i][j]:
            sent = ''.join([text_mystem[i][number]['text'] for number in range(len(text_mystem[i]))]).replace('\s', '')
            print(f'Неснятая омонимия: {lemmas_text_mystem[i][j]}, снятая: {lemmas_disamb_mystem[i][j]}, контекст: {sent}')
```

**Результат**
Снятая дизамбигуация помогает избавиться от контекстной ошибки "стоять"-"стоить", правильно находит лемму "стартап" (при неснятой - "стартапа"), "много" (при неснятой - "больше").

Однако и при снятой дизамбигуации есть ошибки: например, в предложении "пользоваться им может каждый" "им" лемматизируется как "они" (при неснятой верно, "он"), неснятая омонимия: "мочь", снятая: "может" и др.

В целом, значительной разницы между снятой и неснятой омонимией не наблюдается.
